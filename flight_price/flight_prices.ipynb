```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_excel('Data_Train.xlsx')
df_train.head()
```

```python
df_test = pd.read_excel('Test_set.xlsx')
df_test.head()
```

```python
df = pd.concat([df_train, df_test], axis = 0, ignore_index=True)
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df['Date_of_Journey'].str.split("/").str[0]
```

```python
#Featue Engineering
df['Day'] = df['Date_of_Journey'].str.split("/").str[0].astype(int)
df['Month'] = df['Date_of_Journey'].str.split("/").str[1].astype(int)
df['Year'] = df['Date_of_Journey'].str.split("/").str[2].astype(int)
df.head()
```

```python
df.info()
```

```python
df.drop('Date_of_Journey', axis = 1, inplace =True)
```

```python
df.head()
```

```python
df['Arrival_Time'].str.split(" ")
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(" ")[0])
df.head()
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(":").str[0].astype(int)
df['Arrival_minute'] = df['Arrival_Time'].str.split(":").str[1].astype(int)
df.drop('Arrival_Time', axis = 1, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['Dep_Time'].str.split(" ")
```

```python
df['Dep_hour'] = df['Dep_Time'].str.split(":").str[0].astype(int)
df['Dep_minute'] = df['Dep_Time'].str.split(":").str[1].astype(int)
df.drop('Dep_Time', axis = 1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3,  '4 stops':4})
```
